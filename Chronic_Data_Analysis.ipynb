```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("kidney_disease.csv")
df
```

```python
df.isnull().sum()
```

```python
df['age'].plot.box()
df['age'] = df['age'].fillna(df['age'].median())
df['age'].median()
print(df.age.skew())
```

```python
df.info()
```

```python
df['bp'] = df.bp.fillna(df.bp.median())
df['bp'].isnull().any()
```

```python
df.columns
df.drop('id', axis = 1, inplace = True)
```

```python
df.columns = ['age', 'blood_pressure', 'specific_gravity', 'albumin', 'sugar', 'red_blood_cells', 'pus_cell',
              'pus_cell_clumps', 'bacteria', 'blood_glucose_random', 'blood_urea', 'serum_creatinine', 'sodium',
              'potassium', 'haemoglobin', 'packed_cell_volume', 'white_blood_cell_count', 'red_blood_cell_count',
              'hypertension', 'diabetes_mellitus', 'coronary_artery_disease', 'appetite', 'peda_edema',
              'aanemia', 'class']
```

```python
df['packed_cell_volume'] = pd.to_numeric(df['packed_cell_volume'], errors = 'coerce')
df['packed_cell_volume']
```

```python
df['packed_cell_volume'].unique()
```

```python
df['packed_cell_volume'].isnull().sum()
```

```python
df.columns
```

```python
df['white_blood_cell_count'] = pd.to_numeric(df['white_blood_cell_count'], errors = 'coerce')
df[df['white_blood_cell_count'].isnull()]
```

```python
cat_cols = [col for col in df.columns if df[col].dtype == 'object']
cat_cols
```

```python
num_cols = [col for col in df.columns if df[col].dtype != 'object']
num_cols
```

```python
for col in df.columns:
  if df[col].dtype == 'object':
    print(f"{col} has these {df[col].unique()} values\n\n")
```

```python
df['diabetes_mellitus ']  = df['diabetes_mellitus'].replace(to_replace = {' yes' : 'yes','\tyes' : 'yes', '\tno' : 'no' })
df['diabetes_mellitus ']
```

```python
df['coronary_artery_disease'] = df['coronary_artery_disease'].replace('\tno', 'no')
df['coronary_artery_disease'].unique()
```

```python
df['class'].unique()
df['class'] = df['class'].map({'ckd':1, 'not  ckd' : 0, 'ckd\t' : 1})
```

```python
plt.figure(figsize = (10,7))
sns.histplot(df['age'], kde = True, bins = 20, color = 'skyblue')
plt.title('Age of people in kidney disease')
plt.xlabel('Age')
plt.ylabel('Total Count')
plt.show()
```
